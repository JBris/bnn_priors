```python
import numpy as np
import torch
from matplotlib import pyplot as plt

from bnn_priors.data import CIFAR10Augmented, CIFAR10
```

## Inspect data

```python
data = CIFAR10()
```

```python
fig, axs = plt.subplots(ncols=10, figsize=(20,2))

for i, ax in enumerate(axs):
    img = data.norm.train_X[i].numpy().transpose((1,2,0))
    img -= img.min(axis=(0,1))
    img /= img.max(axis=(0,1))

    ax.imshow(img)
    ax.axis("off")
```
